```python
#事前補助金最適化モデル=（額は１タイプ）事後的に１タイプの補助金を与えるモデルとも言える

import FuncDesigner as fd
from openopt import LP
import numpy as np
from numpy.random import *

a,b = 0.01,0.02    #a＝BLの確率、b＝SHの確率

SL = 0.2
BL = 0.4
SH = 0.6
BH = 0.8

# 変数の定義
s,p1,p2,p3,p4 = fd.oovars(5)
# 目的関数
f = s
# 初期解
startPoint = {s:0,p1:0, p2:0, p3:0,p4:0}
constraints = [] 
constraints.append(b*(BH-p1+s)+(1-b)*(BH-p2+s)>=b*(-p4)+(1-b)*(BH-p3+s))
constraints.append(b*(-p4)+(1-b)*(BL-p3+s)>=b*(BL-p1+s)+(1-b)*(BL-p2+s))
constraints.append((1-a)*(p1+s-SH)+a*p4>=(1-a)*(p2+s-SH)+a*(p3+s-SH))
constraints.append((1-a)*(p2+s-SL)+a*(p3+s-SL)>=(1-a)*(p1+s-SL)+a*p4)
constraints.append(b*(BH-p1+s)+(1-b)*(BH-p2+s)>=0)
constraints.append(b*(-p4)+(1-b)*(BL-p3+s)>=0)
constraints.append((1-a)*(p1+s-SH)+a*p4>=0)
constraints.append((1-a)*(p2+s-SL)+a*(p3+s-SL)>=0) 
constraints.append(p1 >= 0)   
constraints.append(p2 >= 0)
constraints.append(p3 >= 0)   
#constraints.append(p4 >= 0)


# 式の定義
m = LP(f, startPoint, constraints=constraints, goal='min',iprint = -1)  
#最小化を補助金で行う,iprintで結果のみを表示

# 解
r = m.solve('cvxopt_lp')
s_opt,p1_opt,p2_opt,p3_opt,p4_opt = r(s,p1,p2,p3,p4)
    
print s_opt,p1_opt,p2_opt,p3_opt,p4_opt    #(9)が成り立つ時、補助金はマイナスの値で出る＝仲介料が取れる
# 2sが補助金支出
```
